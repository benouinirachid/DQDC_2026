# Data Collection & Web Scraping with Python

### What is Data Collection & Preparation?
Data collection is the process of gathering raw information from various sources (websites, APIs, databases, files). Data preparation (or *data wrangling*) transforms that raw data into a clean, structured format ready for analysis, visualization, or machine learning.

**Why it matters:**
- Raw data is often messy, inconsistent, or incomplete.
- The quality of your insights is directly tied to the quality of your input data.
- Automating collection saves hours of manual copying/pasting.

### Data Types & Sources
| Type | Structure | Examples | Collection Method |
|------|-----------|----------|-------------------|
| **Structured** | Tabular, fixed schema | CSV, Excel, SQL tables | `pandas.read_*()`, SQL queries |
| **Semi-Structured** | Hierarchical, flexible | JSON, XML, NoSQL | `json` module, `requests`, `ElementTree` |
| **Unstructured** | No predefined format | HTML, PDFs, images, text | Web scraping (`BeautifulSoup`, `Scrapy`), OCR |

### Essential Python Libraries
- `requests`: Sends HTTP requests to fetch web pages or API responses.
- `BeautifulSoup` (`bs4`): Parses HTML/XML and navigates the DOM tree.
- `pandas`: Structures, cleans, and exports tabular data.
- `json` / `xml.etree.ElementTree`: Handles semi-structured formats.

### The Web Scraping Workflow
1. **Inspect**: Use browser DevTools (F12) to examine HTML structure.
2. **Fetch**: Send an HTTP `GET` request to retrieve the page source.
3. **Parse**: Load HTML into `BeautifulSoup` using a parser (`html.parser`, `lxml`, or `html5lib`).
4. **Extract**: Use `find()`, `find_all()`, or CSS selectors to target elements.
5. **Clean & Structure**: Strip whitespace, handle missing values, convert types.
6. **Export**: Save to CSV, JSON, or database for downstream analysis.
