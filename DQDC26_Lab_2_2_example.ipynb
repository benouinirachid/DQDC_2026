## Parse HTML

````python
from bs4 import BeautifulSoup

html = """
<html>
  <head>
    <title>Book Store</title>
  </head>
  <body>
    <h1 class="title">Welcome to Book Store</h1>
    <ul id="books">
      <li class="book">
        <a href="/book/1">Python Crash Course</a>
        <span class="price">$29.99</span>
      </li>
      <li class="book">
        <a href="/book/2">Automate the Boring Stuff</a>
        <span class="price">$34.99</span>
      </li>
      <li class="book featured">
        <a href="/book/3">Fluent Python</a>
        <span class="price">$49.99</span>
      </li>
    </ul>
  </body>
</html>
"""

soup = BeautifulSoup(html, "html.parser")
# or use the faster lxml parser:
soup = BeautifulSoup(html, "lxml")

print(type(soup))
# <class 'bs4.BeautifulSoup'>

print(soup.title)
# <title>Book Store</title>

print(soup.title.text)
# Book Store
````

## Find elements

Use find() / find_all() / select() to locate the element.

````python
# Returns the FIRST matching tag
first_book = soup.find("li")
print(first_book.text.strip())
````

````python
# Returns ALL matching tags as a list
all_books = soup.find_all("li")
for b in all_books:
    print(b.find("a").text)
````

````python
# Find by CSS class using class_= keyword
prices = soup.find_all("span", class_="price")
for p in prices:
    print(p.text)
````

## Attributes & text

````python
book = soup.find("li", class_="book")

# Get inner text
link = book.find("a")
print(link.text)
# Python Crash Course

# Get an attribute
print(link["href"])
# /book/1

# .get() is safer (returns None if missing)
print(link.get("href"))
# /book/1

# Loop all books and extract data
for item in soup.find_all("li", class_="book"):
    name = item.find("a").text
    price = item.find("span", class_="price").text
    href = item.find("a")["href"]
    print(f"{name}  {price}  {href}")

# Python Crash Course     $29.99  /book/1
# Automate the Boring Stuff  $34.99  /book/2
# Fluent Python            $49.99  /book/3
````

## CSS selectors

````python
# select() returns a list (like find_all)
# select_one() returns first match (like find)

# By tag
soup.select("li")

# By class
soup.select(".price")

# By id
soup.select("#books")

# Descendant  (li inside #books)
soup.select("#books li")

# Multiple classes
soup.select(".book.featured")

# Direct child
soup.select("ul > li")

# By attribute
soup.select('a[href="/book/2"]')

# Practical: get all book titles
titles = [a.text for a in soup.select("#books a")]
print(titles)
# ['Python Crash Course', 'Automate the Boring Stuff', 'Fluent Python']
````
