## Deletion Methods

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Age":          [19, 26, 14, 39, 45, 52, 36],
    "Buy Again?":   [4,  5,  4,  5,  3,  4,  2],
    "Recommend":    [3,  np.nan, 3, 4, 4, 3, 1],
    "Satisfaction": [4,  5,  3,  np.nan, 4, 4, 1],
})

print(df)
```

```python
print(f"Before: {df.shape}")

# Drop any row with at least one NaN
df_listwise = df.dropna()
print(f"After listwise deletion: {df_listwise.shape}")
df_listwise
```

```python
# Drop only rows where a specific column is NaN
df_subset = df.dropna(subset=["Recommend"])
print(f"After dropping NaN in 'Recommend' only: {df_subset.shape}")
df_subset
```

```python
# In-place
df_copy = df.copy()
df_copy.dropna(how="any", inplace=True)
df_copy
```

```python
import pandas as pd
import numpy as np

# Drop column 'note' which has >30% missing values
df_iot = pd.read_csv(
    "https://raw.githubusercontent.com/kjam/data-cleaning-101/master/data/iot_example_with_nulls.csv",
    na_values=["n/a"]
)

df_iot.head()
```

```python
# % of missing values
df_iot.isnull().mean()
```

```python
# Threshold-based column drop
threshold = 0.30
cols_to_drop = df_iot.columns[df_iot.isnull().mean() > threshold].tolist()
print("Dropping columns:", cols_to_drop)

df_col_dropped = df_iot.drop(columns=cols_to_drop)
print(f"Shape before: {df_iot.shape} → after: {df_col_dropped.shape}")
df_col_dropped.head()
```

## Imputation — Univariate

Constant / Arbitrary Value Imputation

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Embarked": ["S", "C", np.nan, "S", np.nan, "Q"],
    "Fare":     [7.25, 71.28, np.nan, 8.05, 13.0, np.nan],
})

# Fill categorical column with most-common value or a sentinel
df["Embarked_filled"] = df["Embarked"].fillna("No")

# Fill numeric column with 0 (sentinel — not always sensible)
df["Fare_filled_0"] = df["Fare"].fillna(-1)

print(df)
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "Age":    [25, np.nan, 30, 22, np.nan, 35, 28, np.nan],
    "Income": [50000, 120000, np.nan, 45000, 80000, np.nan, 60000, 95000],
    "City":   ["Paris", np.nan, "Lyon", "Paris", np.nan, "Paris", "Lyon", "Paris"],
})

# ── Mean imputation
mean_age = df["Age"].mean()
df["Age_mean"] = df["Age"].fillna(mean_age)
print(f"Mean age: {mean_age:.1f}")

# ── Median imputation
median_income = df["Income"].median()
df["Income_median"] = df["Income"].fillna(median_income)
print(f"Median income: {median_income:,.0f}")

# ── Mode imputation
mode_city = df["City"].mode()[0]
df["City_mode"] = df["City"].fillna(mode_city)
print(f"Mode city: {mode_city}")
```

```python
from sklearn.impute import SimpleImputer
# ── Using sklearn SimpleImputer
imputer_mean   = SimpleImputer(strategy="mean")
imputer_median = SimpleImputer(strategy="median")
imputer_mode   = SimpleImputer(strategy="most_frequent")

df["Age_sklearn"]    = imputer_mean.fit_transform(df[["Age"]])
df["Income_sklearn"] = imputer_median.fit_transform(df[["Income"]])

print("\nResult:")
print(df[["Age", "Age_mean", "Age_sklearn", "Income", "Income_median"]].round(1))
```
