# Step 1: Create the Dataset

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

# Sample dataset with 8 instances
data = {
    'Age': [25, np.nan, 26, 32, 45, 22, np.nan, 38],
    'Pclass': [1, 2, 1, 3, np.nan, 3, 2, 1],
    'Fare': [50, 60, 55, 40, 55, 15, 55, 45],
    'Survived': [0, 1, 1, 0, 1, 0, 1, 1]
}

df = pd.DataFrame(data)
print("Original Data with NaNs:")
print(df)
```

# Manual example

We will use the standard Euclidean distance:

$$d(A, B) = \sqrt{\sum (\text{Feature}_i^A - \text{Feature}_i^B)^2}$$

Here is the manual calculation for **Row 4** ($Age=45, Fare=55, Survived=1$) against the other instances to find its missing **Pclass**.

### **Step 1: Distance Calculations (Simplified)**

| Row | Available Features to Compare | Sum of Squared Differences | Distance ($\sqrt{\text{Sum}}$) |
| :--- | :--- | :--- | :--- |
| **0** | Age(45-25), Fare(55-50), Surv(1-0) | $20^2 + 5^2 + 1^2 = 426$ | **20.64** |
| **1** | Fare(55-60), Surv(1-1) | $5^2 + 0^2 = 25$ | **5.00** |
| **2** | Age(45-26), Fare(55-55), Surv(1-1) | $19^2 + 0^2 + 0^2 = 361$ | **19.00** |
| **3** | Age(45-32), Fare(55-40), Surv(1-0) | $13^2 + 15^2 + 1^2 = 395$ | **19.87** |
| **5** | Age(45-22), Fare(55-15), Surv(1-0) | $23^2 + 40^2 + 1^2 = 2130$ | **46.15** |
| **6** | Fare(55-55), Surv(1-1) | $0^2 + 0^2 = 0$ | **0.00** |
| **7** | Age(45-38), Fare(55-45), Surv(1-1) | $7^2 + 10^2 + 0^2 = 149$ | **12.21** |

---

### **Step 2: Identify the $K=2$ Neighbors**
Based on the simplified distance, the two closest rows to Row 4 are:
1.  **Row 6** (Distance: **0.00**)
2.  **Row 1** (Distance: **5.00**)

### **Step 3: Impute Pclass**
Now we take the `Pclass` values from these two neighbors:
* Row 6 Pclass = **2**
* Row 1 Pclass = **2**

$$\text{Imputed Value} = \frac{2 + 2}{2} = \mathbf{2.0}$$



# Step 2: Initialize the KNN Imputer

```python
# Create KNNImputer instance
# We'll use n_neighbors=2 to find the two most similar rows
imputer = KNNImputer(n_neighbors=2, weights="uniform")
```

# Step 3: Fit and Transform the Data

```python
# Fit and transform the dataset
# Note: KNNImputer returns a NumPy array, so we convert it back to a DataFrame
imputed_array = imputer.fit_transform(df)

df_imputed = pd.DataFrame(imputed_array, columns=df.columns)
```

Step 4: View the Results

```python
print("\nImputed Data (No more NaNs):")
print(df_imputed)
```
